```python
import json
from langchain_upstage import UpstageEmbeddings

api_key = 'up_JQDg7exNaCa9WsUszuo1v3dmWc2LZ'

embeddings = UpstageEmbeddings(
    api_key = api_key,
    model = "embedding-query"
)
```

```python
def embed_content(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as infile:
        data = json.load(infile)
    
    content_data = {}
    
    for idx, entry in data.items():
        content = entry
        embedded_content = embeddings.embed_documents([content])
        
        content_data[idx] = embedded_content
    
    with open(output_file, 'w', encoding='utf-8') as outfile:
        json.dump(content_data, outfile, ensure_ascii=False, indent=4)
```

```python
input_file = 'doc_0.json'
output_file= 'embedded_doc0.json'

embed_content(input_file, output_file)
```

```python
def embed_title_and_content(input_file, output_file_all, output_file_title):
    
    with open(input_file, 'r', encoding='utf-8') as infile:
        data = json.load(infile)
    
    embedded_data = {}
    embedded_title_data = {}
    
    for idx, entry in data.items():
        title = entry.get("title")
        content = entry.get("content")
        
        print(f"🌳 {title}")
        
        embedded_title = embeddings.embed_documents([title])
        embedded_content = embeddings.embed_documents([content])
        print(len(embedded_title))
        
        embedded_data[idx] = {
            "title": embedded_title,
            "content": embedded_content
        }
        
        embedded_title_data[idx] = {
            "title": embedded_title
        }
        
    with open(output_file_all, 'w', encoding='utf-8') as outfile_all:
        json.dump(embedded_data, outfile_all, ensure_ascii=False, indent=4)   
    
    with open(output_file_title, 'w', encoding='utf-8') as outfile_title:
        json.dump(embedded_title_data, outfile_title, ensure_ascii=False, indent=4)
```

```python
input_file = 'doc_1.json'
output_file_all = 'embedded_doc1.json'
output_file_title = 'embedded_title.json'

embed_title_and_content(input_file, output_file_all, output_file_title)
```

```python
def embed_content_article(input_file, output_file):
    
    with open(input_file, 'r', encoding='utf-8') as infile:
        data = json.load(infile)
    
    embedded_article_data = {}
    
    for idx, entry in data.items():
        content = entry.get("content") # List
        
        embedded_articles = []
    
        for article_idx, article in enumerate(content, start=0):
            embedded_article = embeddings.embed_documents([article])
            
            embedded_articles.append({
                'article_no.': article_idx,
                'embedded_content': embedded_article
            })
            
        embedded_article_data[idx] = {
            "content": embedded_articles
        }
        
    with open(output_file, 'w', encoding='utf-8') as outfile:
        json.dump(embedded_article_data, outfile, ensure_ascii=False, indent=4)   
```

```python
input_file = 'doc_2.json'
output_file = 'embedded_doc2.json'

embed_content_article(input_file, output_file)
```
