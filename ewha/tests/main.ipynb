```python
# Read test.csv

import pandas as pd
import re

api_key = 'up_JQDg7exNaCa9WsUszuo1v3dmWc2LZ'
file_path = './testset.csv'

def read_data(data_path):
    data = pd.read_csv(data_path)
    prompts = data['prompts']
    answers = data['answers']
    # returns two lists: prompts and answers
    return prompts, answers

prompts, answers = read_data(file_path)
```

```python
# How to use EwhaQA to get response about ewha.pdf

# Note:
# - The printed messages about "maximum iterations," "update temperature," etc., are part of the EwhaQA process and do not indicate errors.
# - For each questions, it takes about 1-2 minutes to get the response --> for 25 questions, it takes about 40 minutes.

from ewha.ewha_run import EwhaQA #  ⭐ import EwhaQA from ewha.ewha_run


ewha = EwhaQA(api_key=api_key, code=5) # ⭐ create EwhaQA class, and put the api key and code; the code must be 5

responses = []

for idx in range(25): # top 25 questions in test.csv for ewha related questions
    question = prompts[idx]
    
    response = ewha.get_response(question) # ⭐ use ewha.get_response to input question and get response
    
    responses.append(response) # Returns '(A)', '(D)'
    print(response) 
```

```python
from ewha.llms.utils import evaluate_accuracy
evaluate_accuracy(responses, answers)
```
