```python
import pandas as pd
import re

from ewha_pdf import EwhaPDF
from llms.utils import evaluate_accuracy
```

```python
file_path = './testset.csv'

def read_data(data_path):
    data = pd.read_csv(data_path)
    prompts = data['prompts']
    answers = data['answers']
    # returns two lists: prompts and answers
    return prompts, answers

prompts, answers = read_data(file_path)
```

```python
def extract_question(text):
    # Use regex to match and remove only the "QUESTION n)" part, without using capturing groups
    text = re.sub(r"QUESTION\d+\)\s*", "", text)
    
    return text
```

```python
api_key = 'up_JQDg7exNaCa9WsUszuo1v3dmWc2LZ'

ewha = EwhaPDF(api_key=api_key, code=0)
responses = []

for idx in range(25):
    question = prompts[idx]
    question = extract_question(question)
    print(question)
    
    response = ewha.ensemble_equally(question)
    responses.append(response)
    print(response)

evaluate_accuracy(answers, responses)
```
